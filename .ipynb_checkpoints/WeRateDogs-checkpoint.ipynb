## Cleaning Data

### Data Quality and Tidiness Issues
#### Replace 'None' with 'NaN' in the columns doogo, floofer, pupper and puppo
#### The columns doggo, floofer, pupper, puppo should be values in one column, not columns
#### Rename names with lower case that are not actually names
#### Replace words with the '&amp' in the column text
#### Convert Denominator and Numerator from int to float
#### The timestamp column is an object. It has to be a datetime object.
#### The column text can be splitted in two columns, text and url
#### Convert the tweet_id to string
#### Combine Denominator and Numerator in one column
#### Remove columns  in_reply_to_user_id and in_reply_to_status_id

#### Creating copies

```python
# Creating copies
df_tweetcp = df_tweet.copy()
df_cp = df.copy()
df_imgcp = df_img.copy()
```

#### Replace 'None' with 'NaN' in the columns doogo, floofer, pupper and puppo

```python
df_cp.doggo = np.where(df_cp.doggo == 'None' , np.nan, df_cp.doggo)
df_cp.floofer = np.where(df_cp.floofer == 'None' , np.nan, df_cp.floofer)
df_cp.pupper = np.where(df_cp.pupper == 'None' , np.nan, df_cp.pupper)
df_cp.puppo = np.where(df_cp.puppo == 'None' , np.nan, df_cp.puppo)
```

```python
df_cp
```

```python
df_cp.head(10)
```

#### The columns doggo, floofer, pupper, puppo should be values in one column, not columns

```python
# Creates a new column for the dog type instead of 4 columns
df_cp['type'] = df_cp['text'].str.extract('(puppo|pupper|floofer|doggo)', expand=True)
```

```python
# Drop the four columns from the DataFrame
columns = ['doggo', 'floofer', 'pupper', 'puppo']
df_cp = df_cp.drop(columns, axis=1)
```

#### Rename names with lower case that are not actually names

```python
# Initialization of variable.
invalid_names = []

# For loop to find ordinary word and check for every name starting with lowercase.
for index in df_cp.name:
    if index.islower():
        invalid_names.append(index)
    elif index is 'a':
        invalid_names.append(index)

# Filter by unique values
invalid_names = list(set(invalid_names))

# Print to test
invalid_names
```

```python
# For loop to replace the invalid names for the string 'None'
for index in invalid_names:
    df_cp.name.replace(index,"None",inplace = True)
```

```python
# Testing if all invalid names were removed
sum(df_cp.name.isin(invalid_names))
```

#### Replace words with the '&amp' in the column text

```python
df_cp['text'] = df_cp['text'].str.replace('&amp;', '&')
```

```python
# Testing if the words were replaced
df_cp[df_cp.text.str.contains('&amp;')]
```

#### Convert Denominator and Numerator from int to float

```python
# Change rating_numerator and rating_denominator form int to float
df_cp.rating_numerator = df_cp.rating_numerator.astype('float')
df_cp.rating_denominator = df_cp.rating_denominator.astype('float')
```

```python
df_cp.info()
```

#### The timestamp column is an object. It has to be a datetime object.

```python
df_cp.timestamp = pd.to_datetime(df_cp.timestamp)
```

```python
df_cp.info()
```

#### The column text can be splitted in two columns, text and url

```python
# Creates a auxliar list to store the values that we are going to split
aux_list = []
aux_url_list = [] 

# For loop to remove the HTML tags
for index in df_cp.text:
    aux_list.append(' '.join(index.split(" ")[:-1]))
    aux_url_list.append(index.split(" ")[-1])

# Update the source column with aux_ls_sr.
df_cp.text = aux_list
df_cp = pd.concat([df_cp, pd.DataFrame(aux_url_list, columns = ['text_url'])], axis = 1)
```

```python
# Testing
df_cp.head(10)
```

#### Convert the tweet_id to string

```python
# Convert the tweet_id to string
df_cp.tweet_id = df_cp['tweet_id'].astype(str)
df_img.tweet_id = df_cp['tweet_id'].astype(str)
```

```python
df_img.info()
```

#### Combine Denominator and Numerator in one column

```python
df_cp['rating'] = df_cp.rating_numerator/df_cp.rating_denominator
df_cp.head()
```

```python
df_cp.drop(columns=['rating_numerator','rating_denominator'], inplace=True)
```

```python
df_cp.head()
```

#### Remove columns  in_reply_to_user_id and in_reply_to_status_id

```python
df_cp.drop(['in_reply_to_status_id', 'in_reply_to_user_id'], axis=1, inplace=True)
```
